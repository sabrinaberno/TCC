### Etapa 2 — Como iniciar a regressão da porcentagem de euploidia?

## Pré-requisitos para a Etapa 2:
- Para iniciar a regressão da porcentagem de euploidia, preciso de uma **nova coluna alvo (y)** com **valores contínuos**, como por exemplo:

```text
Coluna: Porcentagem_Euploidia
Exemplos: 0.41, 0.92, 0.67, ...
```

- Essa coluna precisam estar **relacionada aos mesmos dados morfocinéticos/morfológicos** usados

#### Requisitos essenciais para que a regressão funcione bem:

1. **Uma coluna alvo contínua**, baseada em dados reais. Exemplos:
   - Mosaicismo detectado: 48%
   - Fração de células euploides: 74%
   - Porcentagem obtida via teste genético real
2. **Sem essa coluna real**, **não é possível treinar um modelo de regressão válido**.  
   Simular valores seria estatisticamente incorreto e resultaria em um modelo sem base científica — ele **não teria o que aprender de verdade**.

## Próximo passo prático

### Pipeline para Criar e Usar a Variável `Porcentagem_Euploidia_Estimada`
#### 1. Criar a variável `Porcentagem_Euploidia_Estimada`
- Utilizar regras baseadas em faixas clínicas e correlações previamente definidas.
- Para cada embrião, somar os percentuais atribuídos a cada variável, considerando as faixas favoráveis e a força de correlação com a ploidia.

#### 2. Adicionar essa coluna à planilha
- A nova coluna será a variável alvo (`y`) da regressão contínua.
- Essa coluna representa uma estimativa da porcentagem de euploidia para cada embrião.

#### 3. Separar os dados
- `X`: todas as variáveis preditoras utilizadas no cálculo (excluindo a coluna estimada).
- `y`: a coluna `Porcentagem_Euploidia_Estimada`.

#### 4. Treinar um modelo de regressão
- Modelos recomendados:
  - `MLPRegressor` (rede neural)
  - `RandomForestRegressor` (floresta aleatória)
  - `GradientBoostingRegressor` (boosting)
- Avaliar o desempenho utilizando métricas:
  - MAE (Erro Absoluto Médio)
  - RMSE (Raiz do Erro Quadrático Médio)
  - R² (Coeficiente de Determinação)

#### 5. Validar os resultados
- Analisar a distribuição dos valores previstos vs. os valores estimados.
- Verificar se o modelo consegue aprender de forma gradual a distinção entre embriões com baixo e alto potencial.

### Etapas complementares (opcionais, mas recomendadas)
- Salvar o modelo treinado em um arquivo `.pkl` para uso futuro.
- Comparar visualmente o desempenho do modelo de regressão com um modelo binário (classificação).
- Criar gráficos interpretáveis para análise:
  - Curva de dispersão (previsto vs. real)
  - Importância das variáveis no modelo
