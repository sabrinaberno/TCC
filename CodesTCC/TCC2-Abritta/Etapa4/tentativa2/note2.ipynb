## C√°lculo e Justificativa das Faixas Morfocin√©ticas

**As faixas cl√≠nicas consideradas como favor√°veis foram definidas com base na literatura (CRUZ et al., 2012) e pela an√°lise dos intervalos entre os tempos morfocin√©ticos. Por exemplo, para a vari√°vel **cc2** (t3 - t2), o valor ideal √© menor que 11,9 horas, indicando divis√£o celular regular. Essa faixa foi definida pela subtra√ß√£o dos valores t√≠picos para t3 e t2, conforme observados em embri√µes saud√°veis. De forma similar, as vari√°veis **cc3**, **t5 - t2** e **s2** tiveram suas faixas calculadas com base nas combina√ß√µes dos tempos individuais, garantindo coer√™ncia com os padr√µes de desenvolvimento embrion√°rio observados em sistemas time-lapse.**

### üîπ cc2 (t3 - t2)

- **t2:** 24,3‚Äì27,9 h

- M√≠nimo: 35,4 - 27,9 = 7,5 h

**Faixa favor√°vel cl√≠nica indicada:** < 11,9 h (CRUZ et al., 2012)  
**Faixa sugerida para IA:** 7,5 h ‚Äì 11,9 h  
> *Ideal fisiol√≥gico antes de haver risco de atraso celular.*

---

### üîπ cc3 (t5 - t3)

# A IA pode recalcular automaticamente os pesos (%)

## üß† Recalibragem de Pesos com Base no Coeficiente de Spearman

A IA pode recalcular automaticamente os pesos (%) de cada vari√°vel com base na magnitude do coeficiente de Spearman, de forma **objetiva, justa e escal√°vel**. Esse processo √© conhecido como **normaliza√ß√£o da import√¢ncia relativa**.

### Como funciona:

Voc√™ j√° tem os coeficientes de Spearman (œÅ) entre as vari√°veis e a ploidia. A l√≥gica √© simples:

1. **Obtemos os valores absolutos** dos coeficientes (porque tanto correla√ß√£o negativa quanto positiva forte indicam relev√¢ncia).

2. **Somamos todos os valores absolutos.**

3. **Calculamos o peso percentual** de cada vari√°vel:

$$
\text{Peso (\%)} = \left( \frac{|\rho_i|}{\sum |\rho|} \right) \times 100
$$
